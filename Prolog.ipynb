```python
%Zad. 5            Zadanie działa na SWISH

:- dynamic ksiazka/3.         % Predykat przechowujący informacje o ksiazkach(ksiazka(Tytul, Autor, Gatunek))
:- dynamic ocena/3.           % Predykat przechowujący informacje o ocenach ksiazek (ocena(Uzytkownik, Ksiazka, Ocena))
:- dynamic preferencje/2.    % Predykat przechowujący informacje o preferencjach użytkowników dotyczacych gatunkow literackich (preferencje(Uzytkownik, Gatunek))

% Fakty o książkach, autor i gatunek
ksiazka('Duma u uprzedzenie', austen, romans).
ksiazka('Marsjanin', weir, science_fiction).
ksiazka('Solaris', lem, science_fiction).
ksiazka('Kot w pustym mieszkaniu', szymborska, poezja).
ksiazka('Wichrowe wzgorza', bronte, romans).
ksiazka('Nic dwa razy', szymborska, poezja).

% Fakty o ocenach użytkowników na temat danych ksiązek w skali 1-5
ocena(uzytkownik1, 'Duma u uprzedzenie', 4).
ocena(uzytkownik1, 'Wichrowe wzgorza', 5).
ocena(uzytkownik1, 'Solaris', 2).
ocena(uzytkownik1, 'Nic dwa razy', 5).
ocena(uzytkownik1, 'Kot w pustym mieszkaniu', 5).
ocena(uzytkownik2, 'Solaris', 4).
ocena(uzytkownik2, 'Marsjanin', 5).
ocena(uzytkownik2, 'Nic dwa razy', 2).



% Fakty o preferencjach użytkowników, dotyczące gatunków ksiazek
preferencje(uzytkownik1, poezja).
preferencje(uzytkownik1, romans).
preferencje(uzytkownik2, science_fiction).



% Reguły

% Sprawdzenie preferencji użytkownika
preferuje(Uzytkownik, Gatunek) :-
    preferencje(Uzytkownik, Gatunek).

% Regula znajduje ksiazki w okreslonym gatunku, które są zgodne z preferencjami danego uzytkownika oraz te, które
% mają ocenę równą lub większą niż 4
polecane_ksiazki(Uzytkownik, Gatunek) :-
    findall(Tytul,
        (
            ksiazka(Tytul, _, Gatunek),
            preferuje(Uzytkownik, Gatunek),
            ocena(Uzytkownik, Tytul, Ocena),
            Ocena >= 4
        ),
        Ksiazki),
    Ksiazki \= [], % Upewnij się, że są jakieś książki do polecenia
    write('Polecane książki: '), write(Ksiazki), nl.

% Dynamiczne dodawanie oceny
dodaj_ocene(Uzytkownik, Ksiazka, Ocena) :-
    retractall(ocena(Uzytkownik, Ksiazka, _)),
    assertz(ocena(Uzytkownik, Ksiazka, Ocena)).

Wyniki:
?-polecane_ksiazki(uzytkownik1, poezja).              wynik: Polecane książki: [Kot w pustym mieszkaniu, Nic dwa razy]
?-polecane_ksiazki(uzytkownik2, science_fiction).     wynik: Polecane książki: [Marsjanin, Solaris]
?-polecane_ksiazki(uzytkownik1, romans).              wynik: Polecane książki: [Duma u uprzedzenie, Wichrowe wzgorza]
?-preferuje(uzytkownik1, romans).                     wynik: true
?-preferuje(uzytkownik1, science_fiction).            wynik: false
?-preferuje(uzytkownik2, science_fiction).            wynik: true
?-preferuje(uzytkownik1, poezja).                     wynik: true
```
